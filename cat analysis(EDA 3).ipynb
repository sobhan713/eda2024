```python
dict(vdf.dtypes)
cat=[keys for keys,values in dict(vdf.dtypes).items() if values=='object']
for i in cat:
    cdf=vdf[i].value_counts()
    values=cdf.values
    keys=cdf.keys()
    cols=['label','count']
    df=pd.DataFrame(zip(keys,values),columns=cols)
    path=r'E:\datascience\vdf'
    df.to_csv(f'{path}\\{i}.csv',index=False)
    df
```

```python
for i in cat[1:]:
    cdf=vdf[i].value_counts()
    values=cdf.values
    keys=cdf.keys()
    cols=['label','count']
    df=pd.DataFrame(zip(keys,values),columns=cols)
    plt.figure(figsize=(8,3))
    plt.bar(df['label'],df['count'],width=0.5)
    plt.title(i)
    plt.xlabel('lable')
    plt.ylabel('count')
    plt.show()
```
