$take$

-- 1st argument: indices

- indices will take some values in array : in a list

- it tries to extract the data

- that values we provideed represents columns or rows??

- axis=0 represents about row

- axis=1 represents columns

- by default axis=0

```python
#case 3 :select 3 columns 2nd, 5th ND 7TH
#using  take
visa_df.take([2,5,7],axis=1)

#the output also behaveas dataframe
```

```python
# case4 slicing
visa_df.take([2:5])
```

```python
**visa_df.iloc[<rows>,<columns>]**

**visa_df.iloc[start:stop, start:stop]**
    
**visa_df.iloc[start:stop:step, start:stop:step]**
    
**visa_df.iloc[[list of rows, [list of cols]]**
```
