```python
#using matplotlib
for i in cat[1:]:
    cdf=vdf[i].value_counts()
    values=cdf.values
    keys=cdf.keys()
    cols=['label','count']
    df=pd.DataFrame(zip(keys,values),columns=cols)
    plt.figure(figsize=(8,3))
    plt.bar(df['label'],df['count'],width=0.5)
    plt.title(f'{i} chart')
    plt.xlabel('lable')
    plt.ylabel('count')
    plt.show()
    
```
