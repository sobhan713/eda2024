```python
path=r'E:\data files\\Visadataset.csv'
vdf=pd.read_csv(path)
vdf1=pd.read_csv(path)
```

```python
cat_column=vdf.select_dtypes(include='object').columns
num_column=vdf.select_dtypes(exclude='object').columns
```

```python
# apply map method to all cat columns
1 WE NEED TO GET UNIQUE labels
2 we need to get a numbers= len unique labels
3 we need to create dictionaru
```

```python
vdf=pd.read_csv(path)
un_labe=sorted(vdf['case_status'].unique())
l1=[i for i in range(len(un_labe))]
d1=dict(zip(un_labe,l1))
```

```python
d1
```

```python
vdf['case_status']=vdf['case_status'].map(d1)
```

```python
vdf
```

```python
for i in cat_column:
    un_labe=sorted(vdf[i].unique())
    l1=[i for i in range(len(un_labe))]
    d1=dict(zip(un_labe,l1))
    vdf[i]=vdf[i].map(d1)
```

```python
from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()
vdf1["case_status"]=le.fit_transform(vdf1['case_status'])
vdf1
```

```python
for i in cat_column:
    vdf1[i]=le.fit_transform(vdf1[i])
    
vdf1
```

```python
**fit and transform**

- fit means develop the logic

- transform mean process the logic

- if we obsereve in map method first we developed dict logic

- then we apply that dict to column

- developig the dict is kind of fit

- apply the transform to the column means we are transforming cat to num

- whenever we are transforming or changing the data we are applying the **fit transform**

- if we not change the data then pnly fit
```
